### Preprocessing - scaling data

```python
# show stats

df_small.describe()
```

```python
# standardize data using StandardScaler

scaler = StandardScaler()  
df_scaled = scaler.fit_transform(df_small)
```

```python
pca = PCA()
pca.fit(df_scaled)
```

```python
pca.explained_variance_ratio_
```

```python
plt.figure(figsize=(10,8))
plt.plot(range(1,18), pca.explained_variance_ratio_.cumsum(), marker = 'o', linestyle = '--')
plt.title('Explained Variance of Components')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
```

```python
pca = PCA(n_components = 7)
```

```python
pca.fit(df_scaled)
```

```python
pca.transform(df_scaled)
```

```python
scores_pca = pca.transform(df_scaled)
```

### 2. The elbow technique

```python
# hide future warnings

import warnings
warnings.filterwarnings('ignore', category=FutureWarning)
```

```python
# fit k-means using transformed data from the pca
wcss = []
for i in range(1,21):
    kmeans_pca = KMeans(n_clusters = i, init = 'k-means++', random_state = 42)
    kmeans_pca.fit(scores_pca)
    wcss.append(kmeans_pca.inertia_)
```

```python
# plot elbow curve

plt.figure(figsize=(10,8))
plt.plot(range(1,21), wcss, marker = 'o', linestyle = '--')
plt.xlabel('Number of Clusters')
plt.ylabel('WCSS')
plt.title('K-means with PCA Clustering')
plt.show()
```

#### I am choosing 4 clusters because that is the largest kink in the curve where the curve changes in steepness significantly

### 3. K-means clustering

```python
# create the k-means object

kmeans_pca = KMeans(n_clusters = 4, init = 'k-means++', random_state = 42) 
```

```python
# fit k-means object to the data

kmeans_pca.fit(scores_pca)
```

```python
# create a new data frame with the original features and add the PCA scores and assigned clusters

df_pca_kmeans = pd.concat([df_small.reset_index(drop = True), pd.DataFrame(scores_pca)], axis = 1)
df_pca_kmeans.columns.values[-3:] = ['Component 1', 'Component 2', 'Component 3']


# last column we add contains the pca k-means clustering labels

df_pca_kmeans['Segment K-means PCA'] = kmeans_pca.labels_
```

```python
# check result

df_pca_kmeans.head()
```

```python
# relabel clusters

df_pca_kmeans['Clusters'] = df_pca_kmeans['Segment K-means PCA'].map({0: '1', 1: '2', 2: '3', 3: '4'})
```

```python
# plot the clusters for Birth rate and Infant mortality rate

x_axis = df_pca_kmeans['Birth rate']
y_axis = df_pca_kmeans['Infant mortality rate']
plt.figure(figsize = (10, 8))
sns.scatterplot(x=x_axis, y=y_axis, hue=df_pca_kmeans['Clusters'])
plt.show()
```

### Discussion
#### These clusters describe a data points ranging from nearly 0 infant mortality rate and low birthrate all the ways to high birth rate and high infant mortality rate. There doesn't seem to be a menaingful delineation between clusters for these two variables.
